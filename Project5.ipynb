### Problem 5) 5 Points

* Find the root of the equation $x^2 - 4x - 7 = 0$ near x = 5 using Newton-Raphson Method 

$$x_1 = x_0 - \frac{f(x_0)}{f'(x_0)}$$
$$.$$
$$.$$
$$.$$
$$x_{n+1} = x_{n} - \frac{f(x_n)}{f'(x_n)}$$ 


````python
def f(x):
  return x**2 - 4*x - 7
def df(x):
  return 2*x - 4
def newton_raphson(x_0, tol=1e-6, max_iter=100):
  for i in range(max_iter):
    x_1 = x_0 - f(x_0) / df(x_0)
    if abs(x_1 - x_0) < tol:
      return x_1
    x_0 = x_1
  raise ValueError("No converge")

x_0 = 5
root = newton_raphson(x_0)
print("Root:", root)
````

    - random

````python
import random

secret_number = random.randint(1, 100)
num_guesses = 0

while True:
    try:
        guess = int(input("Enter number between 0 - 100: "))
        num_guesses += 1

        if guess < secret_number:
            print("Number is low")
        elif guess > secret_number:
            print("Number is high")
        else:
            print(f"You guessed the number {secret_number} in {num_guesses} guesses.")
            break
    except ValueError:
        print("Invalid. Enter a number.")
````

### Problem 8) 10 Points

* Create a random discrete signal x[n] has a size 100. Values of x[n] will be between (-1,1)
* Find x[2n] and x[4n] (Downsample)
* Using x[2n] recover x[n]
 * First fill with zeros
 * Calculate zero x[i] using average of (x[i+1]+x[i-1])/2

````python
import random

x = [random.uniform(-1, 1) for _ in range(100)]

x_2n = [x[i] for i in range(0, len(x), 2)]
x_4n = [x[i] for i in range(0, len(x), 4)]

def recover_signal(downsampled_signal):
  recovered_signal = [0] * len(downsampled_signal) * 2
  recovered_signal[::2] = downsampled_signal
  for i in range(1, len(recovered_signal) - 1, 2):
    recovered_signal[i] = (recovered_signal[i - 1] + recovered_signal[i + 1]) / 2
  return recovered_signal

recovered_x = recover_signal(x_2n)

print(recovered_x)
````

$$cos(x) = \sum_{n=0}^{n}\frac{(-1)^nx^{2n}}{(2n)!}$$

````python
import math
import random

def calculate_e_x(x, tol=1e-6, max_iter=100):
  term = 1
  result = 1
  n = 1
  while abs(term) > tol and n < max_iter:
    term = term * x / n
    result += term
    n += 1
  return result

def calculate_e_minus_x(x):
  result = 0
  for n in range(10):
    term = (-1)**n * x**n / math.factorial(n)
    result += term
  return result

def calculate_sin_x(x, tol=1e-6, max_iter=100):
  term = x
  result = x
  n = 1
  while abs(term) > tol and n < max_iter:
    term = -term * x**2 / ((2*n) * (2*n + 1))
    result += term
    n += 1
  return result

def calculate_cos_x(x, tol=1e-6, max_iter=100):
  term = 1
  result = 1
  n = 1
  while abs(term) > tol and n < max_iter:
    term = -term * x**2 / ((2*n - 1) * (2*n))
    result += term
    n += 1
  return result

for _ in range(20):
  x = random.uniform(-5, 5)
  approx_e_x = calculate_e_x(x)
  exact_e_x = math.exp(x)
  error_e_x = abs(approx_e_x - exact_e_x)

  approx_e_minus_x = calculate_e_minus_x(x)
  exact_e_minus_x = math.exp(-x)
  error_e_minus_x = abs(approx_e_minus_x - exact_e_minus_x)

  approx_sin_x = calculate_sin_x(x)
  exact_sin_x = math.sin(x)
  error_sin_x = abs(approx_sin_x - exact_sin_x)

  approx_cos_x = calculate_cos_x(x)
  exact_cos_x = math.cos(x)
  error_cos_x = abs(approx_cos_x - exact_cos_x)

  print(f"x: {x:.5f}")
  print(f"e^x: Approx: {approx_e_x:.10f}, Exact: {exact_e_x:.10f}, Error: {error_e_x:.10f}")
  print(f"e^-x: Approx: {approx_e_minus_x:.10f}, Exact: {exact_e_minus_x:.10f}, Error: {error_e_minus_x:.10f}")
  print(f"Sin(x): Approx: {approx_sin_x:.10f}, Exact: {exact_sin_x:.10f}, Error: {error_sin_x:.10f}")
  print(f"Cos(x): Approx: {approx_cos_x:.10f}, Exact: {exact_cos_x:.10f}, Error: {error_cos_x:.10f}")
  print()
````
