### **1. Calculator Program**

#### **Objective**: Create a simple calculator that can add, subtract, multiply, and divide two numbers.

#### **Steps**:
1. Ask the user to input two numbers.
2. Ask the user to select an operation (e.g., +, -, *, /).
3. Use `if/elif/else` statements to determine which operation to perform.
4. Perform the selected operation and print the result.
5. Use a `while` loop to allow the user to perform another calculation or quit.

```python
while True:
    number1 = float(input("Enter first number: "))
    number2 = float(input("Enter second number: "))

    operator = input("Enter operator (+, -, *, /): ")

    if operator == "+":
        result = number1 + number2
    elif operator == "-":
        result = number1 - number2
    elif operator == "*":
        result = number1 * number2
    elif operator == "/":
        result = number1 / number2
    else:
        print("Invalid. Try again.")
        continue

    print("Answer=", result)

    choice = input("Do it again? (yes/no): ")
    if choice.lower() != "yes":
        break
```

### **2. Even and Odd Number Classifier**

#### **Objective**: Classify numbers as even or odd from a user-provided list.

#### **Steps**:
1. Ask the user to input a list of numbers (comma-separated).
2. Split the input string into a list of numbers.
3. Use a `for` loop to iterate through the list.
4. Use the modulus operator (`%`) to check if each number is even or odd.
5. Print whether each number is even or odd.

```python
numbers = input("Enter list of numbers seperated by commas: ")
numbers_list = [int(num) for num in numbers.split(",")]

for number in numbers_list:
    if number % 2 == 0:
        print(number, "Even")
    else:
        print(number, "Odd")
```

### **3. Shopping List Organizer**

#### **Objective**: Create a simple shopping list program where users can add, view, and remove items.

#### **Steps**:
1. Initialize an empty list to store shopping items.
2. Use a `while` loop to display a menu (Add, View, Remove, Exit).
3. Use `if/elif/else` to handle the user's selection.
4. For "Add", prompt the user for an item and append it to the list.
5. For "View", print all items in the list.
6. For "Remove", allow the user to remove an item by name.
7. Exit the loop when the user selects "Exit".

```python
shopping_list = []

while True:
    print("\nMenu:")
    print("1. Add")
    print("2. View")
    print("3. Remove")
    print("4. Exit")

    choice = input("Enter choice (1-4): ")

    if choice == "1":
        item = input("Enter item: ")
        shopping_list.append(item)
        print("Item added to list.")
    elif choice == "2":
        if shopping_list:
            print("Shopping list:")
            for item in shopping_list:
                print(item)
        else:
            print("Shopping list empty.")
    elif choice == "3":
        item_to_remove = input("Enter item to remove: ")
        if item_to_remove in shopping_list:
            shopping_list.remove(item_to_remove)
            print("Item deleted")
        else:
            print("Unable to find item in the list.")
    elif choice == "4":
        print("Exiting")
        break
    else:
        print("Invalid. Try again.")
```

### **4. Student Grades Dictionary**

#### **Objective**: Store student grades in a dictionary and classify them into grade categories.

#### **Steps**:
1. Create an empty dictionary to store student names and grades.
2. Use a `while` loop to continuously add students and their grades.
3. Classify each grade into categories: A, B, C, D, or F using `if/elif/else`.
4. Print the student names and their corresponding grade categories.
5. Allow the user to exit by typing "done".

```python
student_grades = {}

while True:
    name = input("Students name (or 'done' to exit): ")
    if name == "done":
        break

    grade = float(input("Students grade: "))

    if grade >= 90:
        category = "A"
    elif grade >= 80:
        category = "B"
    elif grade >= 70:
        category = "C"
    elif grade >= 60:
        category = "D"
    else:
        category = "F"

    student_grades[name] = category

print("Students Grades:")
for name, category in student_grades.items():
    print(f"{name}: {category}")
```

### **5. Word Counter**

#### **Objective**: Count the frequency of each word in a sentence.

#### **Steps**:
1. Ask the user to input a sentence.
2. Split the sentence into a list of words.
3. Use a dictionary to count the frequency of each word.
4. Iterate through the list of words, using `if` to check if the word exists in the dictionary.
5. Print the dictionary of word frequencies.

```python
sentence = input("Enter sentence: ")
words = sentence.split()
word_frequency = {}

for word in words:
    if word in word_frequency:
        word_frequency[word] += 1
    else:
        word_frequency[word] = 1

print("Word Frequency:")
for word, frequency in word_frequency.items():
    print(f"{word}: {frequency}")
```

### **6. Number Guessing Game**

#### **Objective**: Create a simple number guessing game.

#### **Steps**:
1. Generate a random number between 1 and 100 using `random.randint`.
2. Use a `while` loop to ask the user to guess the number.
3. If the guess is too high or too low, print a hint.
4. When the guess is correct, exit the loop and print a congratulatory message.

```python
import random
random_number = random.randint(1, 100)

while True:
    guess = int(input("Guess number between 1 - 100: "))
    if guess == random_number:
        print("Great! You guessed number.")
        break
    elif guess < random_number:
        print("Guess was too low.")
    else:
        print("Guess was too high.")
```


### **7. To-Do List Application**

#### **Objective**: Build a simple to-do list manager.

#### **Steps**:
1. Create an empty list to store tasks.
2. Use a `while` loop to display a menu (Add, View, Remove, Exit).
3. For "Add", append the task to the list.
4. For "View", display all tasks.
5. For "Remove", delete a task by its index.
6. Use `if/else` statements to manage menu choices.

```python
tasks = []

while True:
    print("To-Do List Manager")
    print("1. Add")
    print("2. View")
    print("3. Remove")
    print("4. Exit")

    choice = int(input("Enter choice: "))
    if choice == 1:
        task = input("Enter task: ")
        tasks.append(task)
        print("Task added")
    elif choice == 2:
        print("Your Tasks:")
        for index, task in enumerate(tasks):
            print(f"{index + 1}. {task}")
    elif choice == 3:
        index = int(input("Enter task index to remove: ")) - 1
        if 0 <= index < len(tasks):
            del tasks[index]
            print("Task removed")
        else:
            print("Invalid index")
    elif choice == 4:
        print("Exiting")
        break
    else:
        print("Invalid choice")
```

### **8. FizzBuzz Challenge**

#### **Objective**: Implement the classic FizzBuzz problem.

#### **Steps**:
1. Use a `for` loop to iterate over a range of numbers.
2. Use `if/elif/else` to check if a number is divisible by 3, 5, or both.
3. Print "Fizz" for multiples of 3, "Buzz" for multiples of 5, and "FizzBuzz" for multiples of both.

```python
for number in range(1, 101):
    if number % 3 == 0 and number % 5 == 0:
        print("FizzBuzz")
    elif number % 3 == 0:
        print("Fizz")
    elif number % 5 == 0:
        print("Buzz")
    else:
        print(number)
```

### **9. Temperature Converter**

#### **Objective**: Convert temperatures between Celsius, Fahrenheit, and Kelvin.

#### **Steps**:
1. Ask the user to input a temperature and the unit they want to convert from and to.
2. Use `if/elif/else` to handle different conversion formulas.
3. Print the converted temperature.

```python
def C_F(celsius):
    return (celsius * 9/5) + 32
def C_K(celsius):
    return celsius + 273.15
def F_C(fahrenheit):
    return (fahrenheit - 32) * 5/9
def F_K(fahrenheit):
    return (fahrenheit - 32) * 5/9 + 273.15
def K_C(kelvin):
    return kelvin - 273.15
def K_F(kelvin):
    return (kelvin - 273.15) * 9/5 + 32

while True:
    temperature = float(input("Enter temperature: "))
    from_unit = input("Enter unit to convert from (C, F, K): ")
    to_unit = input("Enter unit to convert to (C, F, K): ")

    if from_unit == "C" and to_unit == "F":
        converted_temperature = C_F(temperature)
    elif from_unit == "C" and to_unit == "K":
        converted_temperature = C_K(temperature)
    elif from_unit == "F" and to_unit == "C":
        converted_temperature = F_C(temperature)
    elif from_unit == "F" and to_unit == "K":
        converted_temperature = F_K(temperature)
    elif from_unit == "K" and to_unit == "C":
        converted_temperature = K_C(temperature)
    elif from_unit == "K" and to_unit == "F":
        converted_temperature = K_F(temperature)
    else:
        print("Invalid units")
        continue



    print(f"{temperature} {from_unit} = {converted_temperature} {to_unit}.")
```

### **10. Contact Book**

#### **Objective**: Create a contact book where users can add, remove, and search for contacts.

#### **Steps**:
1. Use a dictionary to store contacts, where the key is the contact name and the value is the phone number.
2. Use a `while` loop to display a menu (Add, Remove, Search, View, Exit).
3. For "Add", ask the user for a name and phone number and add them to the dictionary.
4. For "Remove", allow the user to remove a contact by name.
5. For "Search", display the phone number for a given contact.
6. For "View", display all contacts.


```python
contacts = {}

while True:
    print("Contact Book")
    print("1. Add Contact")
    print("2. Remove Contact")
    print("3. Search Contact")
    print("4. View Contacts")
    print("5. Exit")

    choice = int(input("Enter choice: "))

    if choice == 1:
        name = input("Enter name: ")
        phone = input("Enter phone number: ")
        contacts[name] = phone
        print("Contact added")
    elif choice == 2:
        name = input("Enter name to remove: ")
        if name in contacts:
            del contacts[name]
            print("Contact removed")
        else:
            print("Could not find contact")
    elif choice == 3:
        name = input("Enter name to search: ")
        if name in contacts:
            print(f"{name}: {contacts[name]}")
        else:
            print("Could not find contact")
    elif choice == 4:
        print("Your Contacts:")
        for name, phone in contacts.items():
            print(f"{name}: {phone}")
    elif choice == 5:
        print("Exiting")
        break
    else:
        print("Invalid choice.")
```
