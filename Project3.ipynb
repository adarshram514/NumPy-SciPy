3. Write a Python program that takes a dictionary and creates two lists (A,B). The List ```A```  elements will be keys and the list ```B``` elements will be values. 

```python
# Example
### input ###
{'Beth': [10, 90], 'Dan': [70, 60], 'Kathy': [50, 40], 'JohnDoe': [0, 0]} # Dictionary C
### output ###
Name=['Beth','Dan','Kathy',"JohnDoe"] # This is A
Exam=[[10,90],[70,60],[50,40],[0,0]]  # This is B
```

````python
def create_dictionary_from_lists(list_a, list_b):
  if len(list_a) != len(list_b):
    raise ValueError("Lists must have the same length.")

  return dict(zip(list_a, list_b))

def create_dictionary_from_multiple_lists(list_a, list_b, list_c, list_d):
  if len(list_a) != len(list_b) or len(list_a) != len(list_c) or len(list_a) != len(list_d):
    raise ValueError("Lists must have the same length.")

  return {key: [b, c, d] for key, b, c, d in zip(list_a, list_b, list_c, list_d)}

def create_lists_from_dictionary(dictionary):
  return list(dictionary.keys()), list(dictionary.values())

Name=['Beth','Dan','Kathy',"JohnDoe"]
Exam=[[10,90],[70,60],[50,40],[0,0]]

dictionary_c = create_dictionary_from_lists(Name, Exam)
print(dictionary_c)

Name=['John','Albert',"Layla"]
Exam=[[121,110],[424,515],[620,701]]
HW=[[12,10],[44,55],[60,70]]
Projects=[[2,1],[4,5],[6,7]]

dictionary_z = create_dictionary_from_multiple_lists(Name, Exam, HW, Projects)
print(dictionary_z)

key_list, value_list = create_lists_from_dictionary(dictionary_c)
print(key_list)
print(value_list)
````

```python
# Example
```

````python
def create_dictionary(names, Exam1, Exam2):
    EEClass = {}
    for i in range(len(names)):
        EEClass[names[i]] = [Exam1[i], Exam2[i]]
    return EEClass

def update_exam1_grades(EEClass):
    for key, value in EEClass.items():
        value[0] += 10

def calculate_exam_average(EEClass):
    for key, value in EEClass.items():
        average = (value[0] + value[1]) / 2
        value.append(average)

names = ['Beth', 'Dan', 'Kathy', 'Mark', 'Mary', 'Susie']
Exam1 = [10, 90, 70, 60, 50, 40]
Exam2 = [90, 80, 90, 30, 90, 70]

EEClass = create_dictionary(names, Exam1, Exam2)
print("Original EEClass:", EEClass)

update_exam1_grades(EEClass)
print("After Updating Exam1 Grades:", EEClass)

calculate_exam_average(EEClass)
print("Calculated Averages:", EEClass)
````

## QUESTION 3 (10 Points): 

Create a shopping list file using python

* Ask user his/her name
* Ask shopping date (try to get this date from OS if you can)
* Ask user the list of the items for shopping. User must enter all items at once separated by commas without spaces.
* Please see the example below

```python
# Assume user entered 'Bananas, Apples, Oranges, Bread, Milk'
# file name "Shopping.txt"
Name: John Doe
Date: 9/3/2022
++++++++++++++
Item 1: Bananas 
Item 2: Apples 
Item 3: Oranges 
Item 4: Bread
Item 5: Milk
++++++++++++++
Number of Items: 6
++++ DONE ++++
```


````python
import datetime

name = input("Enter your name: ")

today = datetime.date.today()

shopping_items = input("Enter shopping items separated by commas: ")

file_name = f"Shopping.txt"

with open(file_name, "w") as f:
    f.write(f"Name: {name}\n")
    f.write(f"Date: {today.strftime('%Y-%m-%d')}\n")
    f.write("++++++++++++++\n")

    for item in shopping_items.split(","):
        f.write(f"Item {shopping_items.split(',').index(item) + 1}: {item}\n")

    f.write("++++++++++++++\n")
    f.write(f"Number of Items: {len(shopping_items.split(','))}\n")
    f.write("++++ DONE ++++")

print(f"Shopping list saved as {file_name}")
````

## QUESTION 4 (20 Points): 

Using the shopping list file ```Shopping.txt``` in Question 3.

* Open the file using Python
* Ask user the quantity of the items they want
* Update the file and save it as ```Shopping_Updated.txt```
* Please see the example below

```python
# Assume user entered 'Bananas, Apples, Oranges, Bread, Milk'
# file name "Shopping.txt"
Name: John Doe
Date: 9/3/2022
++++++++++++++
Item 1: Bananas (3 lb)
Item 2: Apples (4 lb)
Item 3: Oranges (4 lb)
Item 4: Bread (2 loaf)
Item 5: Milk (1 gallon)
++++++++++++++
Number of Items: 6
++++ DONE ++++
```

````python
def update_shopping_list(file_name):
  with open(file_name, "r") as f:
    shopping_list = f.readlines()

  item_quantities = {}
  for line in shopping_list:
    if line.startswith("Item"):
      item_name = line.split(":")[1].strip()
      item_quantities[item_name] = None

  print("Enter the quantity for each item:")
  for item, _ in item_quantities.items():
    quantity = input(f"  - {item}: ")
    item_quantities[item] = quantity

  updated_list = []
  for line in shopping_list:
    if line.startswith("Item"):
      item_name = line.split(":")[1].strip()
      updated_list.append(f"Item {line.split()[1]}: {item_name} ({item_quantities[item_name]})")
    else:
      updated_list.append(line)

  new_file_name = "Shopping_Updated.txt"
  with open(new_file_name, "w") as f:
    f.writelines(updated_list)

  print(f"Shopping list updated and saved as {new_file_name}")

update_shopping_list("Shopping.txt")
````

## QUESTION 5 (40 Points): 

Create a student grade calculator and store all entered values in ```Grades.txt``` 

* Ask user his/her name
* Ask user student number
* Ask user to enter HW grades (separated by commas) 
* Ask user to enter Lab grades (separated by commas)
* Ask user to enter Midterm Exam grade
* Ask user to enter Final Exam grade
* Ask user to enter Final Project grade
* Using the grade distribution below calculate student's numerical grade
* Using university letter grade, calculate the letter grade
* Ask if there is anymore students (do this until termination letter is entered) 
* Please see the example below
> Use following grade distribution
> * HW: 20 (use average of all HW)
> * Labs: 30 (use average of all Labs)
> * Midterm Exam: 20 
> * Final Exam: 20 
> * Final Project: 10 

```python
# Assume there are two students 
# file name "Grades.txt"
# Overall grade is not correct value. 
Name: John Doe
Number: 1234
HW: 12,14,23,78,99,100
Labs: 12,14,23,78,99,100
Midterm Exam: 88
Final Exam: 72
Final Project: 88
Overall Grade: 72
Overall Letter Grade: C

    
Name: Jen Scott
Number: 3344
HW: 12,14,23,78,99,100
Labs: 12,14,23,78,99,100
Midterm Exam: 88
Final Exam: 72
Final Project: 88
Overall Grade: 82
Overall Letter Grade: B
```

````python
def calculate_grade(hw_grades, lab_grades, midterm_exam, final_exam, final_project):

  hw_average = sum(hw_grades) / len(hw_grades)
  lab_average = sum(lab_grades) / len(lab_grades)

  overall_grade = (hw_average * 0.2) + (lab_average * 0.3) + (midterm_exam * 0.2) + (final_exam * 0.2) + (final_project * 0.1)

  return overall_grade

def get_letter_grade(numerical_grade):

  if numerical_grade >= 90:
    return "A"
  elif numerical_grade >= 80:
    return "B"
  elif numerical_grade >= 70:
    return "C"
  elif numerical_grade >= 60:
    return "D"
  else:
    return "F"

def main():

  file_name = "Grades.txt"

  while True:
    name = input("Enter the student's name (or 'quit' to exit): ")
    if name.lower() == "quit":
      break

    student_number = input("Enter the student's number: ")

    hw_grades = [int(grade) for grade in input("Enter HW grades separated by commas: ").split(",")]
    lab_grades = [int(grade) for grade in input("Enter Lab grades separated by commas: ").split(",")]
    midterm_exam = int(input("Enter Midterm Exam grade: "))
    final_exam = int(input("Enter Final Exam grade: "))
    final_project = int(input("Enter Final Project grade: "))

    overall_grade = calculate_grade(hw_grades, lab_grades, midterm_exam, final_exam, final_project)
    letter_grade = get_letter_grade(overall_grade)

    with open(file_name, "a") as f:
      f.write(f"Name: {name}\n")
      f.write(f"Number: {student_number}\n")
      f.write(f"HW: {', '.join(map(str, hw_grades))}\n")
      f.write(f"Labs: {', '.join(map(str, lab_grades))}\n")
      f.write(f"Midterm Exam: {midterm_exam}\n")
      f.write(f"Final Exam: {final_exam}\n")
      f.write(f"Final Project: {final_project}\n")
      f.write(f"Overall Grade: {overall_grade:.2f}\n")
      f.write(f"Overall Letter Grade: {letter_grade}\n\n")

if __name__ == "__main__":
  main()
````
